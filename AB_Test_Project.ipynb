```python
df.drop(columns=['Duration (in seconds)','Location Latitude', 'Location Longitude'], inplace=True)
```

```python
from scipy.stats import ttest_ind
columns_to_test = df.select_dtypes(include='number').columns.tolist()
control_group = df[df['Group'] == 'Control']
treatment_group = df[df['Group'] == 'Treatment']

t_test_results = {}

# Perform t-test and calculate means for each column
for column in columns_to_test:
    # Calculate means
    control_mean = control_group[column].mean()
    treatment_mean = treatment_group[column].mean()
    
    # Perform t-test
    t_stat, p_value = ttest_ind(control_group[column], treatment_group[column])
    
    # Store results in dictionary
    t_test_results[column] = {
        'control_mean': control_mean,
        'treatment_mean': treatment_mean,
        't_statistic': t_stat,
        'p_value': p_value
    }

# Display t-test and mean results
for column, result in t_test_results.items():
    print(f"T-test and Mean results for column '{column}':")
    print(f"  - Control Mean: {result['control_mean']:.4f}")
    print(f"  - Treatment Mean: {result['treatment_mean']:.4f}")
    print(f"  - T-statistic: {result['t_statistic']:.4f}")
    print(f"  - P-value: {result['p_value']:.4f}")
    print("\n")
```
